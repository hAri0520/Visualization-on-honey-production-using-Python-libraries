### <font color = "#A020F0"> 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.<font>

```python
sns.scatterplot(data, x='numcol', y='prodvalue')
plt.xlabel('Number of colonies')
plt.ylabel('Value of production')
plt.title('Relationship between number of colonies and value of production')
plt.show()
```

##### Observation

From the above plot we can see that with increase in the number of colonies the value of production of honey also increasesand so w can say that they both are positively correlated.

### <font color = "#A020F0"> 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings. <font>

```python
plt.figure(figsize=(20,10))
sns.boxplot(data, x='year', y='prodvalue')
plt.xlabel('Year')
plt.ylabel('Value of production')
plt.title("How is the Value of production each year")
plt.show()
```

##### Observation

From the above plots we can see that the average cost of production of honey is least in the years **_2000_** and **_2001_** and it is most in the years **_2004_** and **_2008_**

### <font color = "#A020F0"> 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'stocks', 'price per lb', 'prodvalue') <font>

```python
corr_data= data[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']]

sns.heatmap(corr_data.corr(), annot= True, cmap=sns.color_palette("viridis", as_cmap=True))
```

##### observations

The correlation matrix shos the relation ship between two variables. The value of the correlation coefficient lies between -1 and 1  
where:  
-1 indicates a perfectly negative linear correlation between two variables  
0 indicates no linear correlation between two variables  
1 indicates a perfectly positive linear correlation between two variables The further away the correlation coefficient is from zero, the stronger the relationship between the two variables  

For example:  
From the above correlation matrix the following inferences can be made  
1 The number of colonies and the total production of honey is positively correlated, that is as the as the number of colonies increases there is a 95% chance of increases in the total production of honey.  
  
2 The number of colonies and the stock of honey held by producers is positively correlated, that is as the as the number of colonies increases there is a 83% chance that the honey stocked by producer increases.  
  
3 The number of colonies and the total value of production of honey is positively correlated, that is as the as the number of colonies increases there is a 91% chance of increases in the total value of production of honey.  
  
4 The yeild per colony and the proce per lb of the honey is negtively correlated, that is as the yeild per colony increases there is a 36% chance that the proce per l b of honey decreases.
